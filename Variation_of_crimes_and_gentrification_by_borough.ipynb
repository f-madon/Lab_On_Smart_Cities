```python
path = '/content/gdrive/MyDrive/Smart Cities - PROGETTO/Dataset/Crimes/'
df_2019 = pd.read_csv(path + 'df_2019.csv', low_memory=False)
df_2013 = pd.read_csv(path + 'df_2013.csv', low_memory=False)
```

aggiungo i borough al df del gentrification index

```python
df_gentrification['borough'] = df_gentrification['lsoa_name'].str[:-5]
```

```python
df_gentrification.head()
```

```python
#check
print(len(df_gentrification['borough'].unique()))
print(len(df_2019['borough'].unique()))
print(len(df_2013['borough'].unique()))
```

```python
df_2019.drop(['Unnamed: 0'], axis = 1, inplace = True)
df_2019['year'] = '2019'
df_2019.head()
```

```python
df_2013.drop(['Unnamed: 0'], axis = 1, inplace = True)
df_2013['year'] = '2013'
df_2013.head()
```

```python
print(len(df_2019))
print(len(df_2013))
```

```python
df_variation = df_2019
df_variation = df_variation.append(df_2013,ignore_index=True)
df_variation
```

calcolo totale crimini in entrambi gli anni 

```python
crime_totals_2019 = df_2019.groupby(['borough']).size().reset_index(name='count')
crime_totals_2019.rename(columns={"count": "tot_crime_2019"}, inplace=True)

df_variation = pd.merge(df_variation, crime_totals_2019, how="inner", on=["borough"])
df_variation.head()
```

```python
crime_totals_2013 = df_2013.groupby(['borough']).size().reset_index(name='count')
crime_totals_2013.rename(columns={"count": "tot_crime_2013"}, inplace=True)

df_variation = pd.merge(df_variation, crime_totals_2013, how="inner", on=["borough"])
df_variation.head()
```

calcolo variazione numero di crimini 2019-2013:

((num crimini 2019 - num crimini 2013)/num crimini 2013) * 100

```python
df_variation['%variation_crime'] = ((df_variation['tot_crime_2019'] - df_variation['tot_crime_2013'])/df_variation['tot_crime_2013'])*100
df_variation.head()
```

calcolo mediana indice di gentrificazione

```python
G_index_median = df_gentrification.groupby('borough')['G_index'].agg('median').reset_index(name='median')
G_index_median.rename(columns={"median": "G_index_median"}, inplace=True)

df_variation = pd.merge(df_variation, G_index_median, how="inner", on=["borough"])
df_variation.head()
```

calcolo variazione per ogni LSOA

```python
crime_totLSOA_2019 = df_2019.groupby(['LSOA_code']).size().reset_index(name='count')
crime_totLSOA_2019.rename(columns={"count": "tot_crime_LSOA_2019"}, inplace=True)

df_variation = pd.merge(df_variation, crime_totLSOA_2019, how="inner", on=["LSOA_code"])
df_variation.head()
```

```python
crime_totLSOA_2013 = df_2013.groupby(['LSOA_code']).size().reset_index(name='count')
crime_totLSOA_2013.rename(columns={"count": "tot_crime_LSOA_2013"}, inplace=True)

df_variation = pd.merge(df_variation, crime_totLSOA_2013, how="inner", on=["LSOA_code"])
df_variation.head()
```

```python
df_variation['%var_crime_LSOA'] = ((df_variation['tot_crime_LSOA_2019'] - df_variation['tot_crime_LSOA_2013'])/df_variation['tot_crime_LSOA_2013'])*100
df_variation.head()
```

fine

```python
# importing libraries
import seaborn as sb
  
# use regplot
sb.regplot(x = "G_index_median",
            y = "%variation_crime", 
            ci = None,
            data = df_variation)
```

```python
import scipy as sp
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
matplotlib.use('Agg')
sns.set_theme(style="ticks")

x_data = df_variation['G_index_median']
y_data = df_variation['%variation_crime']
sns.scatterplot(x=x_data, y=y_data)

plt.title('Scatterplot between variation of crime and \n gentrification index in London borough')
plt.xlabel('Median of gentrification index')
plt.ylabel('Percentage of variation crime 2019-2013')

r, p = sp.stats.pearsonr(x=x_data, y=y_data)
ax = plt.gca() # Get a matplotlib's axes instance
plt.text(.05, .8, "Pearson's r ={:.3f}".format(r), transform=ax.transAxes)

# The following code block adds the correlation line:
import numpy as np
m, b = np.polyfit(x_data, y_data, 1)
X_plot = np.linspace(ax.get_xlim()[0],ax.get_xlim()[1],100)
plt.plot(X_plot, m*X_plot + b, '-')

plt.savefig('Scatterplot with Pearson r.png', bbox_inches='tight', dpi=300)
plt.close()
```
